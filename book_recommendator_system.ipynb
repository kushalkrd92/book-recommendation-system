<h2> Popularity Based Recommdendator System

```python
ratings_with_name = ratings.merge(books, on='ISBN')
```

```python
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df
```

```python
# Convert 'Book-Rating' to numeric, coercing errors to NaN
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')

# Drop rows where 'Book-Rating' is NaN (due to coercion of non-numeric values)
ratings_with_name_cleaned = ratings_with_name.dropna(subset=['Book-Rating'])

# Calculate the mean rating for each book title on the cleaned data
avg_rating_df = ratings_with_name_cleaned.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'Avg-Rating'},inplace=True)

# Display the result
display(avg_rating_df)
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df.head()
```

```python
popular_df = popular_df[popular_df['Book-Rating'] >= 250].sort_values('Avg-Rating', ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Book-Rating', 'Avg-Rating', 'Image-URL-M']]
```

```python
popular_df.head()
```

```python
popular_df.rename(columns={'Image-URL-M':'Image-URL'},inplace=True)
```

```python
popular_df = popular_df.reset_index(drop=True)
```

```python
popular_df.head()
```

```python
import pickle

pickle.dump(popular_df,open('popular.pkl','wb'))
```

<h2> Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
x.head()
```

```python
users_list = x[x].index
users_list
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(users_list)]
filtered_rating.head()
```

```python
filtered_rating.shape
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
y.head()
```

```python
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings.head()
```

```python
final_ratings.shape
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.head()
```

```python
pt.fillna(0,inplace=True)
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
similarity_scores.shape
```

```python
pt.index[0]
```

```python
pt.index[0][0]
```

```python
sorted(list(enumerate(similarity_scores[1])), key=lambda x:x[1], reverse=True)[1:9]
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:9]      # not selecting 0 as it its own index

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Publisher'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Year-Of-Publication'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('1984')
```

```python
recommend("The Handmaid's Tale")
```
